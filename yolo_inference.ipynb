```python
from ultralytics import YOLO
import torch
import json
import os
# 检查CUDA是否可用，并选择设备
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# 初始化模型并加载到相应设备
model = YOLO('yolov8x')
model.to(device)

# 在选择的设备上运行预测
result = model.predict('input_video/08fd33_4.mp4', save=True, device=device)

# 打印预测结果
# print(result[0])
print('************************')
for box in result[0].boxes:
    print(box)
```

```python
boxes =[]
for frame in result:
    frame_boxes = []
    for box in frame.boxes:
        xyxy = box.xyxy[0]
        frame_boxes.append({
            'x1': xyxy[0].item(),
            'y1': xyxy[1].item(),
            'x2': xyxy[2].item(),
            'y2': xyxy[3].item(),
            'confidence': box.conf.item(),
            'class': box.cls.item()
        })
    boxes.append(frame_boxes)

# 保存到JSON文件
out_dir = 'output_data'
output_file = f'{out_dir}/boxes.json'
if out_dir not in os.listdir():
    os.mkdir(out_dir)
with open(output_file, 'w') as f:
    json.dump(boxes, f, indent=4)

print(f"Box data saved to {output_file}")
```
